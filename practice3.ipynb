```python
DB13 = redis.Redis(host='43.251.253.107', port=1500, db=13)
DB14 = redis.Redis(host='43.251.253.107', port=1500, db=14)
```

```python
#copy one DB to another
for key in DB13.scan_iter():
  key = key.decode('utf-8')
  links_list = DB13.smembers(key)
  for link in links_list:
    link = link.decode('utf-8')
    DB14.sadd(key,link)  
```

```python
# read the contents of a txt file and make a list
with open('/content/advancedWebRanking.txt', 'r') as B:
    links_list = [line.strip() for line in B]
len(links_list)
links_list
```

```python
# write list to pkl file
import pickle
open_file = open('/content/try.pkl', "wb")
pickle.dump(links_list, open_file)
open_file.close()
```

```python
# read/print pkl file
open_file = open('/content/try.pkl', "rb")
loaded_list = pickle.load(open_file)
open_file.close()
print(len(loaded_list))
print(loaded_list)
```

```python
# write contents directory
with open('/content/try.txt','w+') as R:
# extract complete url except 'www'
  for urls in links_list:
    cleanUrl = tldextract.extract(urls)
    if len(cleanUrl.subdomain) == 0:
      linkss = cleanUrl.domain+'.'+cleanUrl.suffix
    elif  cleanUrl.subdomain[0:3] == 'www':
      cu = cleanUrl.subdomain+'.'+cleanUrl.domain+'.'+cleanUrl.suffix
      linkss = cu[4:]
    else:
      linkss = cleanUrl.subdomain+'.'+cleanUrl.domain+'.'+cleanUrl.suffix
    R.write('%s\n'%linkss)
    
```
