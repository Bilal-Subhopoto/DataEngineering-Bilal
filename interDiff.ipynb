```python
!pip install redis
!pip install tldextract
from googlesearch import search
import tldextract
import time
import random
import redis
```

```python
masterDB = redis.Redis(host='43.251.253.107', port=1500, db=11)
DB14 = redis.Redis(host = '43.251.253.107', port = 1500, db = 14)
```

```python
# masterDB domains's list
master_list_dom = []
keys=masterDB.keys()
for key in keys:
  dlink=key.decode("utf-8")
  link=tldextract.extract(dlink)
  domain = link.domain
  domain = str(domain)
  master_list_dom.append(domain)
print(len(master_list_dom))
```

```python
# DB14(google results) keys's list
DB14_keys = []
keys=DB14.keys()
for key in keys:
  dkey=key.decode("utf-8")
  DB14_keys.append(dkey)
print(len(DB14_keys))
```

```python
# Python program to illustrate the intersection
# of two lists using set() method
def intersection(lst1, lst2):
    return list(set(lst1) & set(lst2))

print(len(intersection(master_list_dom, DB14_keys)))
print(intersection(master_list_dom, DB14_keys))
```

```python
# python program to show difference between two lists
master_list_dom = set(master_list_dom)
DB14_keys = set(DB14_keys)

print(len(master_list_dom))
print(len(master_list_dom - DB14_keys))
print(master_list_dom - DB14_keys)
```
